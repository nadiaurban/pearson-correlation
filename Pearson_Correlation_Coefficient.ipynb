### What is the Pearson Correlation Coefficient?

The **Pearson correlation coefficient** (denoted as ùëü) is a statistical measure that evaluates the **linear relationship** between two variables. It quantifies how well the variations in one variable can be predicted by the variations in another.

- **Range**: The value of the Pearson coefficient ranges from -1 to 1:
  - **ùëü = 1**: Perfect positive correlation (both variables increase together).
  - **ùëü = -1**: Perfect negative correlation (one variable increases as the other decreases).
  - **ùëü = 0**: No linear correlation between the variables.

The **p-value** associated with the Pearson coefficient tests the **statistical significance** of the correlation:
- A small p-value (typically < 0.05) indicates that the observed correlation is statistically significant.
- A larger p-value suggests that the observed correlation may have occurred by chance.

This coefficient is widely used in various fields like social sciences, economics, and health research to identify and quantify relationships between variables.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Synthetic_Democracy_Dataset.csv')
data.head()
```

### Box plot creation:

This code creates a **box plot** to visualize the relationship between **Trust in Government** and **Social Equality Support**. A box plot is useful for showing the distribution of data and identifying any outliers.

#### What the code does:
- **Creates a box plot** to compare the distribution of **Trust in Government** and **Social Equality Support**.
- **Visualizes** the spread and central tendency of each variable along with potential outliers.


```python
# Boxplot for Trust in Government vs. Social Equality Support
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Trust_in_Government'], y=data['Social_Equality_Support'], color='lightcoral')
plt.title('Boxplot: Trust in Government vs Social Equality Support', fontsize=16)
plt.xlabel('Trust in Government', fontsize=12)
plt.ylabel('Social Equality Support', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Pearson Correlation in Pandas:

#### Purpose:
This line calculates the **Pearson correlation coefficient** (ùëü) between the variables `Trust_in_Government` and `Social_Equality_Support` using Pandas' `.corr()` method.

#### Result:
The output is a single numeric value representing the **Pearson correlation coefficient**.

#### Interpretation:
The value of ùëü ranges from -1 to 1:

- **ùëü > 0**: A positive relationship exists, meaning as one variable increases, the other tends to increase.
- **ùëü < 0**: A negative relationship exists, meaning as one variable increases, the other tends to decrease.
- **ùëü = 0**: No linear relationship exists between the variables.

In this case, **ùëü = 0.364**, which indicates a **moderate positive linear relationship** between trust in government and support for social equality.


```python
import pandas as pd

# Calculate the Pearson correlation coefficient between Trust_in_Government and Social_Equality_Support
correlation = data["Trust_in_Government"].corr(data["Social_Equality_Support"])
correlation
```

### Pearson correlation coefficient in scipy

#### Purpose:
This code computes both the **Pearson correlation coefficient** (ùëü) and the **p-value** for the relationship between `Trust_in_Government` and `Social_Equality_Support` using Scipy's `pearsonr` function.

#### Result:
- **Pearson Correlation Coefficient (ùëü)**: Measures the strength and direction of the linear relationship.
- **P-value**: Indicates the statistical significance of the correlation.

#### Interpretation:

- **Correlation Coefficient (ùëü)**:
  - **ùëü = 0.364**: There is a moderate positive linear relationship between trust in government and social equality support.

- **P-value (ùëù)**:
  - The extremely small p-value **(ùëù ‚âà 9.81 √ó 10‚Åª¬≥¬≥)** shows the correlation is statistically significant **(ùëù < 0.05)**. This means the observed relationship is unlikely to be due to random chance.

#### Key Takeaway:
By using Scipy, you gain additional insight into whether the relationship is statistically significant, not just its strength and direction.


```python
from scipy.stats import pearsonr

# Calculate Pearson correlation and p-value using scipy
# Calculate Pearson correlation and p-value using scipy
correlation, p_value = pearsonr(data["Trust_in_Government"], data["Social_Equality_Support"])

# Print out results with explanations
print(f"Pearson Correlation Coefficient (r): {correlation:.3f}")
print(f"P-value: {p_value:.3e}")

# Interpretation
if p_value < 0.05:
    print("The correlation is statistically significant (p < 0.05).")
else:
    print("The correlation is not statistically significant (p >= 0.05).")

if correlation > 0:
    print("There is a positive relationship: as Trust in Government increases, Social Equality Support tends to increase.")
elif correlation < 0:
    print("There is a negative relationship: as Trust in Government increases, Social Equality Support tends to decrease.")
else:
    print("There is no linear relationship between Trust in Government and Social Equality Support.")

```
